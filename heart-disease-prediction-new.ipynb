```python
import numpy as np
import pandas as pd
from tensorflow import keras
import tensorflow as tf
from keras.layers import Dense,BatchNormalization,Dropout
from keras import Sequential
from keras import layers
from sklearn.metrics import accuracy_score
```

```python
dset = pd.read_csv("/kaggle/input/heart-disease-prediction-using-logistic-regression/framingham.csv")
dset
```

```python
dset.info()
```

```python
dset = dset.astype({
    'male':float,
    'age':float,
    'currentSmoker':float,
    'prevalentStroke':float,
    'diabetes':float,
    'totChol':float
})
dset.isnull().sum()
```

```python
dset['education']=dset['education'].fillna(dset['education'].mean())
dset['cigsPerDay']=dset['cigsPerDay'].fillna(dset['cigsPerDay'].mean())
dset['BPMeds']=dset['BPMeds'].fillna(dset['BPMeds'].mean())
dset['totChol']=dset['totChol'].fillna(dset['totChol'].mean())
dset['BMI']=dset['BMI'].fillna(dset['BMI'].mean())
dset['heartRate']=dset['heartRate'].fillna(dset['heartRate'].mean())
dset['glucose']=dset['glucose'].fillna(dset['glucose'].mean())
dset['TenYearCHD']=dset['TenYearCHD'].fillna(dset['TenYearCHD'].mean())
dset.isnull().sum()
```

```python
X = dset.drop(["TenYearCHD"],axis=1)
Y = dset["TenYearCHD"]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.20, random_state = 2)
shape = [xtrain.shape[1]]
```

```python
model = Sequential()
model.add(BatchNormalization(input_shape = shape))
model.add(Dense(32,activation='relu'))
model.add(Dropout(rate=0.3))
model.add(Dense(1,activation='sigmoid'))
model.add(Dropout(rate=0.3))

```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=100,batch_size=10)
```

```python
pred = model.predict(xtest)
acc = accuracy_score(ytest,np.round(pred))
print("Accuracy:",acc)
```
