# Heart Beat Normal/Abnormal Comparison

```python
import os
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from time import time
from tqdm import tqdm
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
normal0 = pd.read_csv("/kaggle/input/heartbeat/ptbdb_normal.csv",header=None)
abnormal0 = pd.read_csv("/kaggle/input/heartbeat/ptbdb_abnormal.csv",header=None)
display(normal0)
display(abnormal0)
```

```python
n_cols=len(normal0.columns.tolist())
print(n_cols)
```

```python
cols=[]
for i in range(n_cols):
    cols+=['c'+str(i).zfill(3)]
print(cols)
normal0.columns=cols
abnormal0.columns=cols
```

```python
data1=normal0.iloc[0:200,:]
fig=make_subplots(specs=[[{"secondary_y":False}]])
for i in range(10):
    fig.add_trace(go.Scatter(x=data1.index,y=data1.iloc[i,:],name='No.'+str(i)),secondary_y=False,)
    
fig.update_layout(autosize=False,width=800,height=500,title_text="Normal")
fig.update_xaxes(title_text="cols")
fig.update_yaxes(title_text="value",secondary_y=False)
fig.show()
```

```python
data1=abnormal0.iloc[0:200,:]
fig=make_subplots(specs=[[{"secondary_y":False}]])
for i in range(10):
    fig.add_trace(go.Scatter(x=data1.index,y=data1.iloc[i,:],name='No.'+str(i)),secondary_y=False,)  
fig.update_layout(autosize=False,width=800,height=500,title_text="Abnormal")
fig.update_xaxes(title_text="cols")
fig.update_yaxes(title_text="value",secondary_y=False)
fig.show()
```

```python
normal2=pd.DataFrame([normal0.mean()])
abnormal2=pd.DataFrame([abnormal0.mean()])
mean_df=pd.concat([normal2,abnormal2],axis=0)
diff=[]
for i in range(len(cols)):
    diff+=[abs(mean_df.iloc[0,i]-mean_df.iloc[1,i])]
diff_df=pd.DataFrame(columns=cols,data=[diff])
display(diff_df)
```

```python
data1=mean_df
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=cols,y=data1.iloc[0,:],name='Normal'),secondary_y=False,)
fig.add_trace(go.Scatter(x=cols,y=data1.iloc[1,:],name='Abormal'),secondary_y=False,)  
fig.add_trace(go.Scatter(x=cols,y=diff,name='Difference'),secondary_y=False,)  
fig.update_layout(autosize=False,width=800,height=500,title_text="Mean Value Comparison")
fig.update_xaxes(title_text="cols")
fig.update_yaxes(title_text="value",secondary_y=False)
fig.show()
```
